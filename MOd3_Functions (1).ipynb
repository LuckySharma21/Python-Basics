Q--> What is the difference between a function and a method in Python?

Ans-->Function

A function is a block of reusable code that is defined independently, outside of any class.

You call it directly by its name, and it works on the values you pass as arguments.

Example of a Function:

def add(a, b):

    return a + b

print(add(5, 3))   # Output: 8

Method

A method is also a function, but it is defined inside a class.

It always works on an object (instance of a class) and usually takes self as its first parameter, which refers to the object itself.

You call a method using an object.

Example of a Method:

class Calculator:

    def add(self, a, b):   # method
        return a + b

calc = Calculator()
print(calc.add(5, 3))   # Output: 8

Q-->Explain the concept of generators in Python and how they are defined

Ans-->A generator is a special type of iterator in Python.

Instead of storing all values in memory at once (like lists), a generator produces values on the fly, one at a time.

This makes them memory efficient and useful for large datasets.

Defined using:

1)yield keyword inside a function

2)Generator expression â†’ (x*x for x in range(5))

Example : def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

for num in count_up_to(3):
    print(num)

Q-->What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans-->
1)map()

Purpose: Apply a function to every element of an iterable and return a new sequence of results.

Input: Function + iterable(s)

Output: Iterator (map object)

Example --nums = [1, 2, 3, 4]

squared = map(lambda x: x*x, nums)

print(list(squared))  # [1, 4, 9, 16]

2)filter()

Purpose: Select elements from an iterable that satisfy a condition.

Input: Function that returns True or False + iterable

Output: Iterator (filtered elements)

Example - nums = [1, 2, 3, 4]

evens = filter(lambda x: x % 2 == 0, nums)

print(list(evens))  # [2, 4]

3)reduce() (from functools module)

Purpose: Apply a function cumulatively to the elements of an iterable to reduce it to a single value.

Input: Function + iterable

Output: Single value

Example - from functools import reduce

nums = [1, 2, 3, 4]

sum_all = reduce(lambda x, y: x + y, nums)

print(sum_all)  # 10

Q--> Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

Ans-->

```python
def sum_of_evens(numbers):

    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

nums = [10, 15, 20, 25, 30]
print(sum_of_evens(nums))
```

Q--> Implement a Python function that takes a list of integers and returns a new list containing the squares of
each number.

Ans-->

```python
def square_list(numbers):

    squares = []
    for num in numbers:
        squares.append(num ** 2)
    return squares

# Example usage:
nums = [1, 2, 3, 4, 5]
print(square_list(nums))  # Output: [1, 4, 9, 16, 25]
```

Q--> Write a Python function that checks if a given number is prime or not from 1 to 200.

Ans-->

```python
def is_prime(n):

    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example usage:
for num in range(1, 201):
    if is_prime(num):
        print(num, end=" ")
```

Q--> Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
from functools import reduce

fib = lambda n: reduce(lambda x, _: x + [x[-1] + x[-2]], range(n-2), [0, 1])[:n] if n > 0 else []

print(fib(10))
```

Q-->Write a generator function in Python that yields the powers of 2 up to a given exponent.

Ans-->

```python
def powers_of_two(n):

    for i in range(n + 1):
        yield 2 ** i

# Example usage:
for val in powers_of_two(5):
    print(val, end=" ")
```

Q-->Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

Ans-->

```python

celsius = [0, 20, 37, 100]

fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print("Celsius:", celsius)
print("Fahrenheit:", fahrenheit)
```

Q--> Create a Python program that uses `filter()` to remove all the vowels from a given string.

Ans-->

```python
def remove_vowels(text):
    vowels = "aeiouAEIOU"

    result = ''.join(filter(lambda ch: ch not in vowels, text))
    return result

# Example
string = "Hello World, Python is amazing!"
print("Original:", string)
print("Without vowels:", remove_vowels(string))
```
